```python

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

operation = input("Choose one of the following operator - + * - / % : ")


match operation:
    case "+":
        result = num1 + num2
        print("Result of the operation is", result)
    case "-":
        result = num1 - num2
        print("Result of the operation is", result)
    case "*":
        result = num1 * num2
        print("Result of the operation is", result)
    case "/":
        if num2 == 0:
            print("Invalid input for division")
        else:
            result = num1 / num2
            print("Result of the operation is", result)
    case "%":
        if num2 == 0:
            print("Invalid input for division")
        else:
            result = num1 % num2
            print("Result of the operation is", result)
    case _:
        print("Invalid operator provided")
```
