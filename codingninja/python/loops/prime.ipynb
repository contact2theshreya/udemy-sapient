```python
# prime no check
# compositre number are other than prime number that is >2 factors unlike prime number
N = int(input())
i = 2 # [2, N-1]
isPrime = True
while i <= N-1:
    if N % i == 0:
        # composite number detected
        isPrime = False # marking it not prime
        break

    i += 1

if isPrime:
    print("N was a prime number")
else:
    print("N was a composite number")
print("end")
```

```python
# print factors ,for composite number  u can find factors but for print print -1
N = int(input())

isPrime = True
for i in range(2, N):
    if N % i == 0:
        # we found a factor
        print(i, end = " ")# it make sure cursor goes to same line after 1 space
        isPrime = False

if isPrime:
    print(-1)
```

```python
# reverse of number
n = int(input())

while n > 0:
    rem = n % 10
    print(rem, end = " ")
    n //= 10 #avoid decimal on quoteint while divide by 10
```

```python
n = int(input())
result = 0
while n > 0:
    rem = n % 10
    result = result * 10 + rem# multiply by zero gives result 0 only and when non zero will be added in result ,then leading zero wont come
    n //= 10
print(result)
```
